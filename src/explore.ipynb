```python
from sklearn.preprocessing import LabelEncoder
total_data['y_binary'] = total_data['y'].map({'yes': 1, 'no': 2})
print(total_data['y_binary'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = total_data.drop(['y', 'y_binary'], axis=1)
y = total_data['y_binary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(X_train, X_test, y_train, y_test)
```

```python
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(include=[np.number]).columns.tolist()
```

```python
categorical_features
```

```python
numerical_features
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
X_train_cat = encoder.fit_transform(X_train[categorical_features])
X_test_cat = encoder.transform(X_test[categorical_features])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[numerical_features])
X_test_num = scaler.transform(X_test[numerical_features])
```

```python
X_train_cat_dense = X_train_cat.toarray()
X_test_cat_dense = X_test_cat.toarray()
X_train_processed = np.hstack((X_train_num, X_train_cat_dense))
X_test_processed = np.hstack((X_test_num, X_test_cat_dense))
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_processed, y_train)
```

```python
y_pred = log_reg.predict(X_test_processed)
y_pred_proba = log_reg.predict_proba(X_test_processed)[:, 1]
```

```python
print("Model Performance:")
print("------------------")
print(f"Accuracy: {log_reg.score(X_test_processed, y_test):.4f}")
print("\nClassification Report:")
```
